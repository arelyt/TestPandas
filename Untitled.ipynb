```python
import numpy as np
from math import pi
import datetime as dt
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
tick = pd.read_csv('RIM6_4.trd', usecols=(0, 1, 4, 5, 6))
```

```python
tick = tick.rename(columns={'<DATE>': 'Date', '<TIME>': 'Time',
                            '<LAST>': 'Price', '<VOL>': 'Vol', '<DIRECTION>': 'Dir'})
```

```python
tick.ix[tick.Dir=='Sell', 'VolSell'] = tick.Vol
```

```python
tick.ix[tick.Dir=='Buy', 'VolBuy'] = tick.Vol
```

```python
tr = tick.fillna(0)
```

```python
tick.loc[1110:1120].VolSell.sum() #VolSell.value_counts(dropna=True)
```

```python
Nticks=200 #Количество тиков для бара и ресэмплинг
len(tr.index)
```

```python
# main=pd.DataFrame(index=np.arange(0, (len(tr.index)/Nticks+Nticks)),
#                 columns=('DateTime', 'open', 'high', 'low', 'close', 'vBuy', 'vSell'))
rows_list=[]
for i in range(0, len(tick.index), Nticks):
    dict1={}
#    srez = tr.iloc[i:i+Nticks]
    op=tick.Price.iloc[i]
    hi=tick.Price.iloc[i:i+Nticks].max()
    lo=tick.Price.iloc[i:i+Nticks].min()
    if i+Nticks<len(tick.index):
        cl=tick.loc[i+Nticks-1].Price
    else:
        cl=tick.loc[len(tr.index)-1].Price
    
    vBuy=tick.VolBuy.iloc[i:i+Nticks].sum()
    vSell=tick.VolSell.iloc[i:i+Nticks].sum()
    nBuy=tick.VolBuy.iloc[i:i+Nticks].count()
    nSell=tick.VolSell.iloc[i:i+Nticks].count()
    tClose=tr.Time.iloc[i:i+Nticks].max()
    
    dict1.update({'Time': tClose, 'open': op, 'high': hi,
                 'low': lo, 'close': cl, 'TickBuy': nBuy,
                'VolBuy': vBuy, 'TickSell': nSell, 'VolSell': vSell})
    rows_list.append(dict1)
    
main=pd.DataFrame(rows_list)
    
```
