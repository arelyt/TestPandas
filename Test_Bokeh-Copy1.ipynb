```python
from math import pi
import datetime as dt
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
import numpy as np
```

```python
from bokeh.models import LinearAxis, Range1d
```

```python
bhp = pd.read_csv('RIM6_4.trd', usecols=(0, 1, 4, 5, 6), header=0, 
                  parse_dates={'DateTime': [0, 1]}, index_col=0)
```

```python
bhp = bhp.rename(columns={'<LAST>': 'Price', '<VOL>': 'Vol', '<DIRECTION>': 'Dir'})
```

```python
bhp['VolSell'] = bhp.apply(lambda row: (row['Vol'] if row['Dir']=='Sell' else 0), axis=1)
```

```python
bhp['VolBuy'] = bhp.apply(lambda row: (row['Vol'] if row['Dir']=='Buy' else 0), axis=1)
```

```python
print type(bhp.columns)
```

```python
res = bhp.Price.resample('60s').ohlc()
```

```python
print res
```

```python
resSell = bhp.VolSell.resample('60s').sum()
```

```python
resBuy = bhp.VolBuy.resample('60s').sum()
```

```python
res_new = res.join(pd.DataFrame(resSell))
```

```python
res_new1 = res_new.join(pd.DataFrame(resBuy))
```

```python
print res_new1
```

```python
mids = (res.open + res.close)/2
spans = abs(res.close - res.open)
print res, mids, spans
inc = res.close > res.open
dec = res.open > res.close
```

```python
inc, dec
```

```python
w = 12*60*60
```

```python
p = figure(x_axis_type='datetime', y_range=(90500, 95000), plot_width=1024, plot_height=800, title='RIM6 CandleStick')
p.xaxis.major_label_orientation = pi/4
p.grid.grid_line_alpha = 0.3
p.segment(res.index, res.high, res.index, res.low, color='black')
p.rect(res.index[inc], mids[inc], w, spans[inc], fill_color='#6CF350', line_color='green')
p.rect(res.index[dec], mids[dec], w, spans[dec], fill_color='#FF0000', line_color='red')
p.extra_y_ranges = {"TickSpeed": Range1d(start=0, end=50)}
p.line(res_new1.index, res_new1.VolSell/60, color='blue', y_range_name='TickSpeed', alpha=0.5)
p.line(res_new1.index, res_new1.VolBuy/60, color='red', y_range_name='TickSpeed', alpha=0.5)
p.add_layout(LinearAxis(y_range_name='TickSpeed'), 'left')
show(p)
```
